```python
menu = """

[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> """

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3

while True:

    opcao = input(menu)

    if opcao == "d":
        deposito = int(input("Digite o valor a ser depositado: "))
        extrato += f"Depósito: R$ {deposito:.2f}\n"
        saldo += deposito
        print(f"Seu saldo atual é de R$ {saldo:.2f}")

    elif opcao == "s":
      saque = float(input("Digite o valor a ser sacado: "))
      if saldo - saque <= 0:
        print("Não é possível sacar o dinheiro por falta de saldo")
      elif saque > limite:
        print('O limite para saques é de R$ 500,00')
      elif numero_saques >= LIMITE_SAQUES:
        print('Você já realizou o limite de saques diários')
      else:
        extrato += f"Saque: R$ {saque:.2f}\n"
        numero_saques += 1
        saldo -= saque
        print(f"Seu saldo atual é de R$ {saldo:.2f}")

    elif opcao == "e":
        print(extrato)
        print(f"Seu saldo atual é de R$ {saldo:.2f}")

    elif opcao == "q":
      print('Obrigado por usar nossos serviços, volte sempre!')
      break

    else:
        print("Operação inválida, por favor selecione novamente a operação desejada.")
```
